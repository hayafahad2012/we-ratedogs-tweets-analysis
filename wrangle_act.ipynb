#### Quality issues
1. In `twitter-archive-enhanced` some dogs names are missing or wrong
2. missing values from several columns in `twitter-archive-enhanced.csv`
3. Retweet related columns should be removed 
4. values in `rating_denominator` needs to be all 10
5. Ids fileds should be objects
6. Instead of True or Fales in `image-predictions.tsv` 1 or 0 are a better representation and intrduce more potential.
7. `timestamp` in `df1` and `df2` data type need to be datetime
8. Change `p1_dog` to `first_prediction` and so on. This make columns more descriptive


#### Tidiness issues 

##### Tidy data requirements:
1. Each variable forms a column.
2. Each observation forms a row.
3. Each type of observational unit forms a table.

##### Issues:
- Accourding to 1 dog life-stage should be one column
- Accourding to 2 and 3 all tables should be into one table, since there are only one observational unit which is tweets.

### Define:
Merge dog stages columns into one column

### Code 

```python
df1_clean.doggo.replace('None', '', inplace=True)
df1_clean.floofer.replace('None', '', inplace=True)
df1_clean.pupper.replace('None', '', inplace=True)
df1_clean.puppo.replace('None', '', inplace=True)
```

```python
df1_clean['stage'] = df1_clean.doggo + df1_clean.floofer + df1_clean.pupper + df1_clean.puppo
df1_clean.sample(3)
```

```python
df1_clean.loc[df1_clean.stage == 'doggopupper', 'stage'] = 'doggo,pupper'
df1_clean.loc[df1_clean.stage == 'doggopuppo', 'stage'] = 'doggo,puppo'
df1_clean.loc[df1_clean.stage == 'doggofloofer', 'stage'] = 'doggo,floofer'
```

### Define: 
Add breed and confdence columns to df2_clean

### Code

```python
selections = [(df2_clean['is_prediction_1_dog']==1),(df2_clean['is_prediction_2_dog']==1),(df2_clean['is_prediction_3_dog']==1)]
breeds= [df2_clean['prediction_1'],df2_clean['prediction_2'],df2_clean['prediction_3']]
cofidence=[df2_clean['prediction_1_confident'],df2_clean['prediction_2_confident'],df2_clean['prediction_3_confident']]

df2_clean['breed']= np.select(selections,breeds, default='none')
df2_clean['cofidence']= np.select(selections,cofidence, default=0)
```
